# EDA - Exploratory Data Analysis

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from src.config import ProjectPath 
```

```python
data_name = "Brain"
path = ProjectPath(data_name=data_name)
```

```python
raw_data = path.raw_path
save_dir = str(path.eda_result_dir / "plot")
os.makedirs(save_dir, exist_ok=True)
df = pd.read_csv(raw_data)
```

```python
print(df.shape)
df.info()


# isnull -> bool df have true if that cell is null
# first sum -> all null in a coll
# second sum -> all null in df
total_nans = df.isnull().sum().sum()
print("NaN values: ", total_nans)
```

```python
print(df.iloc[:,1:6].describe())
```

```python
sns.countplot(x="V1", data=df)
plt.title("countplot for analyse the classes")
save_path = os.path.join(save_dir, "countplot.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")
print("󰋩 Saved countplot")
plt.show()
```
