```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
from src.config import ProjectPath
```

```python
data_name = "Brain"

path = ProjectPath(data_name)
```

```python

wrapper_results_dir = path.wrapper_dir 

seeded_metrics_path = Path(f"{wrapper_results_dir}/raw/seededsfsselector/run_20260410_002553_new_suffix_3seeds/metrics/metrics.csv")
sklearn_metrics_path = Path(f"{wrapper_results_dir}/raw/sklearnsfsselector/run_20260410_000220/metrics/metrics.csv")
plot_output_path = Path("results/Brain/evaluation/plots/time_comparison_seeded3_vs_sklearn_brain.png")
plot_output_path.parent.mkdir(parents=True, exist_ok=True)
```

```python
seeded_df = pd.read_csv(seeded_metrics_path)
sklearn_df = pd.read_csv(sklearn_metrics_path)

comparison_df = pd.DataFrame(
    {
        "algorithm": ["SeededSFS (3 seeds)", "SklearnSFS"],
        "total_fit_time_ms": [
            float(seeded_df.loc[0, "total_fit_time_ms"]),
            float(sklearn_df.loc[0, "total_fit_time_ms"]),
        ],
    }
)
comparison_df["total_fit_time_sec"] = comparison_df["total_fit_time_ms"] / 1000
comparison_df
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
fig, ax = plt.subplots(figsize=(9, 5))

bars = ax.bar(
    comparison_df["algorithm"],
    comparison_df["total_fit_time_sec"],
    color=["#2a9d8f", "#e76f51"],
)

ax.set_title("Brain Dataset: Fit Time Comparison", fontsize=14)
ax.set_xlabel("Algorithm")
ax.set_ylabel("Total Fit Time (seconds)")

for bar, value in zip(bars, comparison_df["total_fit_time_sec"]):
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f"{value:.2f}s",
        ha="center",
        va="bottom",
        fontsize=10,
    )

fig.tight_layout()
fig.savefig(plot_output_path, dpi=300, bbox_inches="tight")
plt.show()
print(f"Saved plot to: {plot_output_path}")
```
