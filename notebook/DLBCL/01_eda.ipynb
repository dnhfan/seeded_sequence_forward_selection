# EDA - Exploratory Data Analysis

## lib import

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

from src.config import ProjectPath
```

## reading data

```python
data_name = "DLBCL"

path = ProjectPath(data_name)

today = datetime.now().strftime("%Y-%m-%d")

raw_data = path.raw_path
plot_dir = path.eda_result_dir / "plot"

os.makedirs(plot_dir, exist_ok=True)

df = pd.read_csv(raw_data)
```

explore!

```python
print(df.shape)
df.info()

total_nans = df.isnull().sum().sum()
print("NaN vals: ", total_nans)
```

```python
print(df.iloc[:,0:6].describe())
```

count plot for class distribution

```python
sns.countplot(x="V1", data=df)
plt.title("countplot for analyse the classes")
save_path = os.path.join(plot_dir, "countplot.png")
plt.savefig(save_path, dpi=300, bbox_inches="tight")
print("󰋩 Saved countplot")
plt.show()
```
