```python
import pandas as pd

from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
```

# Trying to select data using sklearn sfs with filterd data union

```python
data_name = "NCI"
data_path = f"data/processed/{data_name}/{data_name}_Union_60features.csv"
df = pd.read_csv(data_path)
```

```python
X = df.iloc[:, 1:]
y: pd.Series = df.iloc[:, 0]
```

```python
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
```

```python
# Using cv=2 instead of cv=4 because class 3 has only 4 samples
# With cv=4, some folds would have only 1 sample from class 3, causing the error
sk_sfs = SequentialFeatureSelector(dt, n_features_to_select=5, n_jobs=-2, scoring='accuracy', cv=5) #type: ignore
sk_sfs.fit(X, y)
```

```python
X_tran = sk_sfs.transform(X)
selected_features = X.columns[sk_sfs.get_support()]

df_new = pd.DataFrame(X_tran, columns=selected_features)

df_new.insert(0, y.name, y.values)

output_path = f"data/processed/{data_name}/{data_name}_SKSFS_5features.csv"
df_new.to_csv(output_path,index=False)
print("ok")
```
