# *Custom Filter Method*

```python
import pandas as pd
from src.filter import FeatureFilter

from src.config import ProjectPath
```

```python
# load data
data_name = "NCI"
path = ProjectPath(data_name)
data_path = path.raw_path
df = pd.read_csv(data_path)
```

run the all filter method 

```python
# split target and features
# y = pd.Series(df["V1"])
# X = df.drop(columns=["V1"])

y = pd.Series(df.iloc[:,0])
X = df.iloc[:,1:]
```

```python
# prepare for filter
valid_method = [
    "variance",
    "correlation",
    "chi_squared",
    "mutual_information",
    "anova_f_test"
    ]

n_features = 50
```

```python
# filter!

print(f" running all filter method...")

for m in valid_method:
    print("-"*10)
    print(f"󰜎 running {m.upper()}")
    print("-"*10)

    # init selector
    selector = FeatureFilter(method=m, n_features=n_features)

    # fit and transform the data
    X_filtered = selector.fit_transform(X, y)

    filter_dir = str(path.filter_dir)

    save_path = f"{filter_dir}/{data_name}_{m}_{n_features}features.csv"

    selector.save_filtered_data(X_filtered, y, save_path)

print("󰄬 DONE!, pls check the processed (filtered) data in data/processed.")
```
