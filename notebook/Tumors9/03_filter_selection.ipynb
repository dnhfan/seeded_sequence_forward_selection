# *Custom Filter Method*

```python
import pandas as pd
from src.filter import FeatureFilter

from datetime import datetime

today = datetime.now().strftime("%Y-%m-%d")
```

```python
# load data
data_name = "Tumors9"
data_path = f"data/raw/{data_name}.csv"
df = pd.read_csv(data_path)
```

run the all filter method 

```python
# split target and features
y = pd.Series(df["V1"])
X = df.drop(columns=["V1"])
```

```python
# prepare for filter
valid_method = [
    "variance",
    "correlation",
    "chi_squared",
    "mutual_information",
    "anova_f_test"
    ]

n_features = 50
```

```python
# filter!

print(f" running all filter method...")

for m in valid_method:
    print("-"*10)
    print(f"󰜎 running {m.upper()}")
    print("-"*10)

    # init selector
    selector = FeatureFilter(method=m, n_features=n_features)

    # fit and transform the data
    X_filtered = selector.fit_transform(X, y)

    save_path = f"data/processed/{data_name}/filter{n_features}/{data_name}_{m}_{n_features}features.csv"

    selector.save_filtered_data(X_filtered, y, save_path)

print("󰄬 DONE!, pls check the processed (filtered) data in data/processed.")
```
