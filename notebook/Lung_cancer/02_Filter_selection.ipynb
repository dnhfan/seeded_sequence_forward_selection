# *Filter Method*

```python
import pandas as pd
from src.filter import FeatureFilter
```

```python

# load data
data_path = "data/processed/Lung_cancer_preprocessed.csv"
df = pd.read_csv(data_path)
```

```python
# split target and features
y = pd.Series(df['V1'])
X = df.drop(columns=['V1'])

print(df.shape)
print(f"X.shape = {X.shape}")
print(f"y.shape = {y.shape}")
```

```python
# prepare for filter !

valid_methods = [
    "variance",
    "correlation",
    "chi_squared",
    "mutual_information",
    "anova_f_test",
]
n_features = 50
```

```python
# filterrr!

print(f" running all filter method...")

for m in valid_methods:
    print("-"*10)
    print(f"󰜎 running {m.upper()}")
    print("-"*10)

    # init selector
    selector = FeatureFilter(method=m,n_features=n_features)

    # learn (fit) and select (transform) features
    X_filtered = selector.fit_transform(X,y)

    save_path = f"data/processed/Lung_cancer/Lung_cancer_{m}_{n_features}features.csv"

    selector.save_filtered_data(X_filtered,y,save_path)
print("󰄬 DONE!, pls check the processed (filtered) data in data/processed.")
```
